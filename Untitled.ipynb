```python
class InstagramTalentScout:
    def __init__(self, username, password, headless=False):
        self.username = username
        self.password = password
        
        # Configure Chrome options
        chrome_options = Options()
        if headless:
            chrome_options.add_argument("--headless")
        chrome_options.add_argument("--no-sandbox")
        chrome_options.add_argument("--disable-dev-shm-usage")
        chrome_options.add_argument("--disable-notifications")
        chrome_options.add_argument("--disable-infobars")
        chrome_options.add_argument("--lang=en-US")
        chrome_options.add_argument("--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.110 Safari/537.36")
        
        # Initialize the Chrome driver
        self.driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)
        self.driver.maximize_window()
        self.wait = WebDriverWait(self.driver, 8)
        self.creators_data = []
        
    def login(self):
        """Login to Instagram"""
        try:
            logger.info("Logging in to Instagram...")
            insta_url = "https://www.instagram.com/" 
            self.driver.get(insta_url)
            
            # Wait for the login page to load and enter credentials
            username_input = self.wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, "input[name='username']")))
            password_input = self.wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, "input[name='password']")))
                
            username_input.clear()
            username_input.send_keys(self.username)
            password_input.clear()
            password_input.send_keys(self.password)
            
            # Click login button
            login_button = self.wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, "button[type='submit']"))).click()
            
            # Handle "Save Your Login Info?" dialog
            try:
                not_now_button = self.wait.until(EC.element_to_be_clickable((By.XPATH, "//div[contains(text(), 'Not now')]"))).click()
                logger.info("Dismissed 'Save Login Info' prompt")
            except TimeoutException:
                logger.info("No 'Save Login Info' prompt appeared")
            
            # Handle notifications dialog
            try:
                not_now_notif = self.wait.until(EC.element_to_be_clickable((By.XPATH, "//div[contains(text(), 'Not Now')]"))).click()
                logger.info("Dismissed notifications prompt")
            except TimeoutException:
                logger.info("No notifications prompt appeared")
                
            logger.info("Successfully logged in to Instagram")
            return True
        except Exception as e:
            logger.error(f"Login failed: {str(e)}")
            return False
    
    def search_hashtag(self, hashtag):
        """Search for posts by hashtag"""
        try:
            logger.info(f"Searching hashtag: #{hashtag}")
            self.driver.get(f"https://www.instagram.com/explore/tags/{hashtag}/")
            
            # Wait for the search results to load
            time.sleep(3)
            self.wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, "article")))
            
            # Get recent posts
            posts = self.wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, "article a")))
            post_urls = [post.get_attribute('href') for post in posts[:20]]  # Get first 20 posts
            
            return post_urls
        except Exception as e:
            logger.error(f"Error searching hashtag: {str(e)}")
            return []
    
    def search_trending_hashtags(self, industry_hashtags=None):
        """Search for trending hashtags in a specific industry"""
        if industry_hashtags is None:
            industry_hashtags = ["viral", "trending", "fyp", "contentcreator", "digitalcreator"]
        
        all_posts = []
        for hashtag in industry_hashtags:
            posts = self.search_hashtag(hashtag)
            if posts:  # Log posts found for each hashtag
                logger.info(f"Found {len(posts)} posts for #{hashtag}")
            else:
                logger.warning(f"No posts found for #{hashtag}")
                
            all_posts.extend(posts)
            time.sleep(random.uniform(2, 5))  # Random delay to avoid rate limiting
        
        return list(set(all_posts))  # Remove duplicates

    # def parse_number(self, text):
    #     if 'K' in text:
    #         return int(float(text.replace('K', '')) * 1000)
    #     if 'M' in text:
    #         return int(float(text.replace('M', '')) * 1000000)
    #     return int(text.replace(',', ''))
    
    def extract_post_data(self, post_url):
        """Extract engagement data from a post"""
        try:
            logger.info(f"Analyzing post: {post_url}")
            self.driver.get(post_url)
            time.sleep(3)
            
            # Extract username
            username = self.wait.until(EC.presence_of_element_located((By.XPATH, "//a[contains(@class, 'notranslate')]"))).text
            
            # Extract view count for videos
            try:
                views_element = self.driver.find_element(By.XPATH, "//span[contains(text(), 'views')]/..")
                views_text = views_element.text
                views = int(re.sub(r'[^\d]', '', views_text))
                has_million_views = views >= 100000
            except NoSuchElementException:
                views = 0
                has_million_views = False

            logger.debug(f"Post views: {views}, Has million views: {has_million_views}")
            
            # Extract likes
            try:
                likes_element = self.driver.find_element(By.XPATH, "//section//span/span")
                likes_text = likes_element.text
                likes = int(re.sub(r'[^\d]', '', likes_text))
            except NoSuchElementException:
                likes = 0

            logger.debug(f"Post likes: {likes}")
            
            # Extract comments count
            try:
                comments_element = self.driver.find_element(By.XPATH, "//span[contains(text(), 'comments')]")
                comments_text = comments_element.text
                comments = int(re.sub(r'[^\d]', '', comments_text))
            except NoSuchElementException:
                comments = 0

            logger.debug(f"Post comments: {comments}")
            
            # Return post data
            return {
                'username': username,
                'post_url': post_url,
                'views': views,
                'likes': likes,
                'comments': comments,
                'has_million_views': has_million_views,
                'engagement_rate': (likes + comments) / (views if views > 0 else 1) * 100
            }
        except Exception as e:
            logger.error(f"Error extracting post data: {str(e)}")
            return None
    
    def analyze_creator_profile(self, username):
        """Analyze a creator's profile for engagement trends"""
        try:
            logger.info(f"Analyzing profile: {username}")
            self.driver.get(f"https://www.instagram.com/{username}/")
            time.sleep(3)
            
            # Get follower count
            followers_element = self.wait.until(EC.presence_of_element_located((By.XPATH, "//a[contains(@href, 'followers')]/span")))
            followers_text = followers_element.text.replace(',', '')
            followers = int(re.sub(r'[^\d]', '', followers_text))
            
            # Get recent posts
            posts = self.wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, "article a")))
            post_urls = [post.get_attribute('href') for post in posts[:10]]  # Last 10 posts
            
            post_data = []
            for url in post_urls:
                data = self.extract_post_data(url)
                if data:
                    post_data.append(data)
                time.sleep(random.uniform(1, 3))
            
            # Calculate average engagement
            if post_data:
                avg_engagement = sum(post['engagement_rate'] for post in post_data) / len(post_data)
                last_post_engagement = post_data[0]['engagement_rate'] if post_data else 0
                
                # Check if on a hot streak (15-20% above average)
                on_hot_streak = last_post_engagement >= avg_engagement * 1.15
                
                return {
                    'username': username,
                    'followers': followers,
                    'avg_engagement_rate': avg_engagement,
                    'last_post_engagement': last_post_engagement,
                    'on_hot_streak': on_hot_streak,
                    'has_million_view_video': any(post['has_million_views'] for post in post_data),
                    'recent_posts': post_data
                }
            return None
        except Exception as e:
            logger.error(f"Error analyzing profile {username}: {str(e)}")
            return None
    
    def find_creators(self, min_followers=500, industry_hashtags=None):
        """Find creators matching our criteria"""
        post_urls = self.search_trending_hashtags(industry_hashtags)
        
        # Extract unique usernames from posts
        usernames = set()
        for url in post_urls:
            try:
                data = self.extract_post_data(url)
                if data:
                    usernames.add(data['username'])
            except Exception as e:
                logger.error(f"Error processing URL {url}: {str(e)}")
            time.sleep(random.uniform(1, 3))
        
        # Analyze each creator's profile
        for username in usernames:
            profile_data = self.analyze_creator_profile(username)
            if profile_data:
                # Relax criteria for hot streak or engagement rate if necessary
                if profile_data and (profile_data['has_million_view_video'] or profile_data['on_hot_streak']) and profile_data['followers'] >= min_followers:
                    self.creators_data.append(profile_data)
                    logger.info(f"Found matching creator: {username}")
                else:
                    logger.debug(f"Skipping {username}: Does not meet criteria (hot streak, million views, or engagement rate)")
            time.sleep(random.uniform(3, 6))
        
        return self.creators_data
    
    def send_message(self, username, message_template):
        """Send a DM to a creator"""
        try:
            logger.info(f"Attempting to message: {username}")
            self.driver.get(f"https://www.instagram.com/{username}/")
            time.sleep(2)
            
            # Click message button
            message_btn = self.wait.until(EC.element_to_be_clickable((By.XPATH, "//div[text()='Message']"))).click()
            time.sleep(2)
            
            # Type message
            message_input = self.wait.until(EC.element_to_be_clickable((By.XPATH, "//textarea[@placeholder='Message...']")))
            message_input.send_keys(message_template)
            time.sleep(1)
            
            # Send message
            send_btn = self.wait.until(EC.element_to_be_clickable((By.XPATH, "//button[text()='Send']"))).click()
            
            logger.info(f"Message sent to {username}")
            time.sleep(random.uniform(5, 10))  # Random delay between actions
            return True
        except Exception as e:
            logger.error(f"Failed to send message to {username}: {str(e)}")
            return False
    
    def reach_out_to_creators(self, message_template=None):
        """Reach out to all identified creators"""
        if message_template is None:
            message_template = "Hi there! I noticed your amazing content, especially your recent viral video. " \
                              "We're looking to collaborate with talented creators like you. " \
                              "Would you be interested in discussing a potential partnership? Thanks!"
        
        results = []
        for creator in self.creators_data:
            username = creator['username']
            success = self.send_message(username, message_template)
            results.append({
                'username': username,
                'message_sent': success,
                'followers': creator['followers'],
                'avg_engagement': creator['avg_engagement_rate'],
                'on_hot_streak': creator['on_hot_streak'],
                'has_million_view': creator['has_million_view_video']
            })
            time.sleep(random.uniform(60, 120))  # Longer delay between messages to avoid rate limits
        
        return results
    
    def export_results(self, filename="creator_outreach_results.csv"):
        """Export results to CSV"""
        if self.creators_data:
            df = pd.DataFrame(self.creators_data)
            df.to_csv(filename, index=False)
            logger.info(f"Results exported to {filename}")
    
    def close(self):
        """Close the browser"""
        self.driver.quit()
        logger.info("Browser closed")
```

```python
# Constants
INSTAGRAM_URL = "https://www.instagram.com"
USERNAME = "soumyajit_1729"
PASSWORD = "Spal@2003"
HASHTAG_TO_SEARCH = ["fyp", "creator", "contentcreator", "viral", "trending"]
ENGAGEMENT_THRESHOLD = 1.15  # 15% above average engagement
MIN_VIEWS_FOR_VIRAL = 1000000  # 1 million views

# Initialize WebDriver
def init_driver():
    chrome_options = webdriver.ChromeOptions()
    chrome_options.add_argument("--headless")
    chrome_options.add_argument("--disable-gpu")
    chrome_options.add_argument("--window-size=1920,1080")
    
    chromedriver_path = r'C:\Users\sanso\Downloads\chromedriver-win64\chromedriver-win64\chromedriver.exe'

    # Create a Service object with the path to chromedriver
    service = Service(executable_path=chromedriver_path)
    
    # Initialize the Chrome WebDriver with the Service object
    driver = webdriver.Chrome(service=service, options=chrome_options)
    driver.implicitly_wait(30)
    # insta_url = "https://www.instagram.com/" 
    # driver.get(insta_url)
    return driver


def random_delay(min_seconds=2, max_seconds=5):
    time.sleep(random.randint(min_seconds, max_seconds))
    

# Login to Instagram
def login(driver):
    driver.get(f"{INSTAGRAM_URL}/accounts/login/")
    random_delay()
    
    try:
        # Enter credentials
        username = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.CSS_SELECTOR, "input[name='username']")))
        password = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.CSS_SELECTOR, "input[name='password']")))
        username.send_keys(USERNAME)
        password.send_keys(PASSWORD)
    
        # Click login button
        # driver.find_element(By.XPATH, '//button[@type="submit"]').click()
        log_in = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.CSS_SELECTOR, "button[type='submit']"))).click()
        random_delay()
        
    
        not_now = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, "//div[contains(text(), 'Not now')]"))).click()
        random_delay()
        
        logger.info("Successfully logged in to Instagram")
        
    except TimeoutException as e:
        logging.error(f"Login failed: {e}")

def search_hashtag(driver, hashtag):
    try:
        logging.info("Waiting for search box...")
        search_box = WebDriverWait(driver, 30).until(
            EC.presence_of_element_located((By.XPATH, '//input[@placeholder="Search"]'))
        )
        logging.info("Search box found. Entering hashtag...")
        search_box.send_keys(hashtag)
        time.sleep(2)

        # Press Enter to search
        search_box.send_keys(Keys.ENTER)
        time.sleep(2)
        search_box.send_keys(Keys.ENTER)
        time.sleep(2)
    except TimeoutException:
        logging.error("Search box not found. Dumping page source for debugging.")
        with open("page_source.html", "w", encoding="utf-8") as f:
            f.write(driver.page_source)
        driver.save_screenshot("screenshot.png")

# Scrape posts from the hashtag page
def scrape_posts(driver):
    posts = driver.find_elements(By.XPATH, '//div[@class="v1Nh3 kIKUG  _bz0w"]')
    post_links = [post.find_element(By.TAG_NAME, 'a').get_attribute('href') for post in posts]
    return post_links

# Scrape engagement data from a post
def scrape_engagement(driver, post_url):
    driver.get(post_url)
    time.sleep(3)

    try:
        # Scrape likes (for images) or views (for videos)
        likes_or_views = driver.find_element(By.XPATH, '//section[@role="button"]/span').text
        if 'views' in likes_or_views.lower():
            views = int(likes_or_views.replace(',', '').replace(' views', ''))
            return {'views': views}
        else:
            likes = int(likes_or_views.replace(',', ''))
            return {'likes': likes}
    except NoSuchElementException:
        return None

# Check if a creator is on a hot streak
def is_hot_streak(engagement_data):
    if len(engagement_data) < 5:
        return False

    # Calculate average engagement
    engagement_rates = [data['likes'] for data in engagement_data if 'likes' in data]
    if not engagement_rates:
        return False

    average_engagement = np.mean(engagement_rates)
    latest_engagement = engagement_rates[-1]

    # Check if latest engagement is above the threshold
    return latest_engagement > average_engagement * ENGAGEMENT_THRESHOLD

# Send a message to a creator
def send_message(driver, username, message):
    driver.get(f"{INSTAGRAM_URL}/{username}/")
    time.sleep(3)

    try:
        # Click the message button
        driver.find_element(By.XPATH, '//div[text()="Message"]').click()
        time.sleep(2)

        # Send the message
        message_box = driver.find_element(By.XPATH, '//textarea[@placeholder="Message..."]')
        message_box.send_keys(message)
        time.sleep(2)
        driver.find_element(By.XPATH, '//button[text()="Send"]').click()
        time.sleep(2)
    except (NoSuchElementException, ElementClickInterceptedException):
        print(f"Could not send message to {username}")
```

```python
# Main function to automate the bot
def automate_bot():
    driver = init_driver()
    login(driver)
    search_hashtag(driver, HASHTAG_TO_SEARCH)
    post_links = scrape_posts(driver)

    for post_link in post_links:
        engagement_data = []
        for _ in range(5):  # Scrape last 5 posts
            data = scrape_engagement(driver, post_link)
            if data:
                engagement_data.append(data)

        if engagement_data and is_hot_streak(engagement_data):
            username = post_link.split("/")[-2]
            message = "Hey, we'd like to hire you! That viral video you did? Make more of those!"
            send_message(driver, username, message)

    driver.quit()

# Run the bot
if __name__ == "__main__":
    automate_bot()
```

```python
class InstagramTalentScout:
    def __init__(self, username, password, headless=False):
        self.username = username
        self.password = password
        
        # Configure Chrome options
        chrome_options = Options()
        if headless:
            chrome_options.add_argument("--headless")
        chrome_options.add_argument("--no-sandbox")
        chrome_options.add_argument("--disable-dev-shm-usage")
        chrome_options.add_argument("--disable-notifications")
        chrome_options.add_argument("--disable-infobars")
        chrome_options.add_argument("--lang=en-US")
        chrome_options.add_argument("--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.110 Safari/537.36")
        
        # Initialize the Chrome driver
        self.driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)
        self.driver.maximize_window()
        self.wait = WebDriverWait(self.driver, 5)
        self.creators_data = []
        
    def login(self):
        """Login to Instagram"""
        try:
            logger.info("Logging in to Instagram...")
            insta_url = "https://www.instagram.com/" 
            self.driver.get(insta_url)
            
            # Wait for the login page to load and enter credentials
            username_input = self.wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, "input[name='username']")))
            password_input = self.wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, "input[name='password']")))
            
            username_input.clear()
            username_input.send_keys(self.username)
            password_input.clear()
            password_input.send_keys(self.password)
            
            # Click login button
            login_button = self.wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, "button[type='submit']"))).click()
            
            # Handle "Save Your Login Info?" dialog
            try:
                not_now_button = self.wait.until(EC.element_to_be_clickable((By.XPATH, "//div[contains(text(), 'Not now')]"))).click()
                logger.info("Dismissed 'Save Login Info' prompt")
            except TimeoutException:
                logger.info("No 'Save Login Info' prompt appeared")
            
            # Handle notifications dialog
            try:
                not_now_notif = self.wait.until(EC.element_to_be_clickable((By.XPATH, "//div[contains(text(), 'Not Now')]"))).click()
                logger.info("Dismissed notifications prompt")
            except TimeoutException:
                logger.info("No notifications prompt appeared")
                
            logger.info("Successfully logged in to Instagram")
            return True
        except Exception as e:
            logger.error(f"Login failed: {str(e)}")
            return False
    
    def search_hashtag(self, hashtag):
        """Search for posts by hashtag"""
        try:
            logger.info(f"Searching hashtag: #{hashtag}")
            self.driver.get(f"https://www.instagram.com/explore/tags/{hashtag}/")
            time.sleep(5)

            try:
                # Wait for posts to be visible
                self.wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, "article")))
            except TimeoutException:
                logger.error(f"No posts found for hashtag #{hashtag}")
                return []

            # # Switch to Recent posts
            # try:
            #     recent_tab = self.wait.until(EC.element_to_be_clickable((By.XPATH, "//div[contains(text(), 'Recent')]"))).click()
            #     logger.info("Successfully Switched to Recent posts")
            #     time.sleep(2)
            # except Exception as e:
            #     logger.info(f"Recent posts tab unavailable: {str(e)}. Using default view")
            #     logger.warning("Recent tab not found, using default posts")

            # Scroll to load more posts
            last_height = self.driver.execute_script("return document.body.scrollHeight")
            scroll_attempts = 0
            max_scrolls = 7

            while scroll_attempts < max_scrolls:
                # Scroll down
                self.driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
                time.sleep(3)  # Wait for content to load
                
                # Calculate new scroll height
                new_height = self.driver.execute_script("return document.body.scrollHeight")
                
                if new_height == last_height:
                    break
                    
                last_height = new_height
                scroll_attempts += 1

            # Find all post links
            post_elements = self.driver.find_elements(By.CSS_SELECTOR, "article a")
            post_urls = []
            
            for element in post_elements:
                try:
                    url = element.get_attribute('href')
                    if url and '/p/' in url:  # Verify it's a post URL
                        post_urls.append(url)
                except:
                    continue
            
            # Remove duplicates while preserving order
            unique_posts = list(dict.fromkeys(post_urls))
            
            # Take most recent posts
            posts_to_analyze = unique_posts[:30]  # Increased from 20 to 30
            
            logger.info(f"Found {len(posts_to_analyze)} unique posts for #{hashtag}")
            return posts_to_analyze
            
        except Exception as e:
            logger.error(f"Error searching hashtag {hashtag}: {str(e)}")
            return []
                
            # Get recent posts
        #     posts = self.wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, "article a")))
        #     post_urls = list([post.get_attribute('href') for post in posts[:20]])  # Get first 20 posts
        #     logger.info(f"Found {len(post_urls)} posts for #{hashtag}")
        #     return post_urls
        # except Exception as e:
        #     logger.error(f"Error searching hashtag: {str(e)}")
        #     return []
    
    
    def search_trending_hashtags(self):
        """Search for trending hashtags in a specific industry"""
        # if industry_hashtags is None:
        industry_hashtags = ["viral", "trending", "fyp", "contentcreator", "digitalcreator"]
        all_posts = []
        
        for hashtag in industry_hashtags:
            try:
                posts = self.search_hashtag(hashtag)
                if posts:  # Log posts found for each hashtag
                    logger.info(f"Found {len(posts)} posts for #{hashtag}")
                    all_posts.extend(posts)
                else:
                    logger.warning(f"No posts found for #{hashtag}")
                
                time.sleep(random.uniform(3, 5))  # Random delay to avoid rate limiting

            except Exception as e:
                logger.error(f"Error processing hashtag #{hashtag}: {str(e)}")
                continue
                
        # Remove duplicates while preserving order
        unique_posts = list(dict.fromkeys(all_posts))
        logger.info(f"Total unique posts found across all hashtags: {len(unique_posts)}")
        
        return unique_posts
    
    def extract_post_data(self, post_url):
        """Extract engagement data from a post"""
        try:
            logger.info(f"Analyzing post: {post_url}")
            self.wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, "header section")))
            self.driver.get(post_url)
            time.sleep(3)
            
            # Extract username
            username_link = self.wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, "header a.notranslate")))
            username = username_link.text
            # username = self.wait.until(EC.presence_of_element_located((By.XPATH, "//a[contains(@class, 'notranslate')]"))).text

            # Determine if post is a video
            is_video = len(self.driver.find_elements(By.CSS_SELECTOR, "svg[aria-label='Video']")) > 0
            
            # Extract view count for videos
            if is_video:
                try:
                    views_element = self.driver.find_element(By.XPATH, "//span[contains(text(), 'views')]/..")
                    views_text = views_element.text
                    views = int(re.sub(r'[^\d]', '', views_text))
                    has_million_views = views >= 100000
                    likes = 0
                except NoSuchElementException:
                    views = 0
                    has_million_views = False
    
                logger.debug(f"Post views: {views}, Has million views: {has_million_views}")
            else:
                # Extract likes
                try:
                    likes_element = self.driver.find_element(By.XPATH, "//section//span/span")
                    likes_text = likes_element.text
                    likes = int(re.sub(r'[^\d]', '', likes_text))
                except NoSuchElementException:
                    likes = 0
                views = 0
                has_million_views = False
    
                logger.debug(f"Post likes: {likes}")
            
            # Extract comments count
            try:
                comments_element = self.driver.find_element(By.XPATH, "//span[contains(text(), 'comments')]")
                comments_text = comments_element.text
                comments = int(re.sub(r'[^\d]', '', comments_text))
            except NoSuchElementException:
                comments = 0

            # Calculate engagement rate based on post type
            engagement_denominator = views if is_video else (likes if likes > 0 else 1)
            engagement_rate = ((likes + comments) / engagement_denominator) * 100 if engagement_denominator else 0

            logger.debug(f"Post comments: {comments}")
            
            # Return post data
            return {
                'username': username,
                'post_url': post_url,
                'views': views,
                'likes': likes,
                'comments': comments,
                'has_million_views': has_million_views,
                'engagement_rate': (likes + comments) / (views if views > 0 else 1) * 100
            }
        except Exception as e:
            logger.error(f"Error extracting post data: {str(e)}")
            return None
    
    def analyze_creator_profile(self, username):
        """Analyze a creator's profile for engagement trends"""
        try:
            logger.info(f"Analyzing profile: {username}")
            self.driver.get(f"https://www.instagram.com/{username}/")
            time.sleep(3)
            
            # Get follower count
            followers_element = self.wait.until(EC.presence_of_element_located((By.XPATH, "//a[contains(@href, 'followers')]/span")))
            followers_text = followers_element.text.replace(',', '')
            followers = int(re.sub(r'[^\d]', '', followers_text))
            logger.info(f"{username} has {followers} followers")
            
            # Get recent posts
            posts = self.wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, "div._aabd > a")))
            post_urls = [post.get_attribute('href') for post in posts[:10]]  # Last 10 posts
            
            post_data = []
            for url in post_urls:
                data = self.extract_post_data(url)
                if data:
                    # Calculate engagement based on followers
                    engagement = (data['likes'] + data['comments']) / followers * 100 if followers > 0 else 0
                    data['engagement_rate'] = engagement
                    post_data.append(data)
                time.sleep(random.uniform(1, 3))
            
            # Calculate average engagement
            if post_data:
                avg_engagement = sum(post['engagement_rate'] for post in post_data) / len(post_data)
                last_post_engagement = post_data[0]['engagement_rate'] if post_data else 0
                on_hot_streak = last_post_engagement >= avg_engagement * 1.10
                has_million_views = any(p['views'] >= 1000000 for p in post_data if p['views'])
                return {
                    'username': username,
                    'followers': followers,
                    'avg_engagement_rate': avg_engagement,
                    'last_post_engagement': last_post_engagement,
                    'on_hot_streak': on_hot_streak,
                    'has_million_view_video': any(post['has_million_views'] for post in post_data),
                    'recent_posts': post_data
                }
            return None
        except Exception as e:
            logger.error(f"Error analyzing profile {username}: {str(e)}")
            return None
    
    def find_creators(self, min_followers=500, min_engagement=0.5):
        """Find creators matching our criteria"""
        # post_urls = self.search_trending_hashtags(industry_hashtags)
        try:
            # Get posts from trending hashtags
            post_urls = self.search_trending_hashtags()
            
            if not post_urls:
                logger.error("No posts found to analyze!")
                return []
            
            logger.info(f"Starting analysis of {len(post_urls)} posts")
            
            # Extract unique usernames from posts
            usernames = set()
            for url in post_urls:
                try: 
                    data = self.extract_post_data(url)
                    if data and data['likes'] > 100:
                        usernames.add(data['username'])
                    time.sleep(1)
                except Exception as e:
                    logger.error(f"Error processing post {url}: {str(e)}")
                    continue
            logger.info(f"Found {len(usernames)} unique creators to analyze")
            
            # Analyze each creator's profile
            for username in usernames:
                try:
                    profile_data = self.analyze_creator_profile(username)
                    if profile_data:
                        # Relax criteria for hot streak or engagement rate if necessary
                        if profile_data and (profile_data['followers'] >= min_followers and profile_data['avg_engagement_rate'] >= min_engagement or profile_data['on_hot_streak']) or profile_data['has_million_view_video']:
                            self.creators_data.append(profile_data)
                            logger.info(f"Qualified creator: @{username}")
                        else:
                            logger.debug(f"Skipping {username}: Does not meet criteria (hot streak, million views, or engagement rate)")
                    time.sleep(random.uniform(2, 3))
                
                except Exception as e:
                    logger.error(f"Error analyzing profile {username}: {str(e)}")
                    continue

            logger.info(f"Successfully identified {len(self.creators_data)} qualifying creators")
            return self.creators_data
                    
        except Exception as e:
            logger.error(f"Error in find_creators: {str(e)}")
            return []
    
    def send_message(self, username, message_template):
        """Send a DM to a creator"""
        try:
            logger.info(f"Attempting to message: {username}")
            self.driver.get(f"https://www.instagram.com/{username}/")
            time.sleep(2)
            
            # Click message button
            message_btn = self.wait.until(EC.element_to_be_clickable((By.XPATH, "//div[text()='Message']"))).click()
            time.sleep(2)
            
            # Type message
            message_input = self.wait.until(EC.element_to_be_clickable((By.XPATH, "//textarea[@placeholder='Message...']")))
            message_input.send_keys(message_template)
            time.sleep(1)
            
            # Send message
            send_btn = self.wait.until(EC.element_to_be_clickable((By.XPATH, "//button[text()='Send']"))).click()
            
            logger.info(f"Message sent to {username}")
            time.sleep(random.uniform(5, 10))  # Random delay between actions
            return True
        except Exception as e:
            logger.error(f"Failed to send message to {username}: {str(e)}")
            return False
    
    def reach_out_to_creators(self, message_template=None):
        """Reach out to all identified creators"""
        if message_template is None:
            message_template = "Hi there! I noticed your amazing content, especially your recent viral video. " \
                              "We're looking to collaborate with talented creators like you. " \
                              "Would you be interested in discussing a potential partnership? Thanks!"
        
        if not self.creators_data:
            logger.warning("No creators found to send messages to!")
            return []
        
        results = []
        for creator in self.creators_data:
            username = creator['username']
            logger.debug(f"Attempting to message: {username}")
            success = self.send_message(username, message_template)
            results.append({
                'username': username,
                'message_sent': success,
                'followers': creator['followers'],
                'avg_engagement': creator['avg_engagement_rate'],
                'on_hot_streak': creator['on_hot_streak'],
                'has_million_view': creator['has_million_view_video']
            })
            time.sleep(random.uniform(60, 120))  # Longer delay between messages to avoid rate limits
        
        return results
    
    def export_results(self, filename="creator_outreach_results.csv"):
        """Export results to CSV"""
        if self.creators_data:
            df = pd.DataFrame(self.creators_data)
            df.to_csv(filename, index=False)
            logger.info(f"Results exported to {filename}")
    
    def close(self):
        """Close the browser"""
        self.driver.quit()
        logger.info("Browser closed")
```
